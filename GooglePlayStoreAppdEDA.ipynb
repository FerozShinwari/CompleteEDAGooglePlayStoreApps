> Let's convert the `M` and `K` units into bytes and then remove the `M` and `K` from the values and convert them into numeric

- >***Important*** : In order to furthure explore the size column, and visualize its insights we need to convert all the entries in one same unit. Therefore we define the below function to make all the values of size column in on single unit. See the function below Which are complelty explained in comments, Hope you will enjoy it.

```python
#converting the whole size into bytes first.
#lets define a function for it.
def convert_size(size):    
    if isinstance(size, str):  #isinstance mean inside the size first deals with string
        if 'k'in size:     #defining our string
            return float(size.replace('k',''))*1024     #we ask to replace k with empy space and make it bytes by *1024
        elif 'M' in size:
            return float(size.replace('M', ''))*1024 *1024    #make it floate coz mb values are  continues 
        elif 'Varies with device' in size:   #if this comes ,make it empty
            return np.nan        #return empty box
        return size       #where there is no mb,kb or varies with device make it as it is.
        
        
```

```python
#now lets apply the above function
df['Size']=df['Size'].apply(convert_size)
```

```python
#now lets check the size value for mb, kb and varies with device
df['Size'] 
```

Only problem is to remove `'+'` sign and comma , and convert the column into an 'int'

- So lets try lambda function and tell python to replace '+' sign form our 'install' column.

But How ???????????????????????????

lets see the code below and everything is explained in comments with it. Hope you enjoy it.

```python
#remove the + and comma,
#so we need to tell the computer if there is + sign replace it with a single free sapce ''.
#otherwise if it is only a number like zero only.
#then make no changes and just print as it is. (else x means that)
df['Installs']=df['Installs'].apply(lambda x: x.replace('+','') if '+'in str(x) else x)

#lets remove the commas now
#We tell python to look into column, locate the comma , change it into a free space ''. otherwise leave it as it is.
df['Installs']=df['Installs'].apply(lambda x: x.replace(',','') if ','in str(x) else x)

# convert the install column to numeric (integers because this is the number of installs/count)
df['Installs'] = df['Installs'].apply(lambda x: int(x))
```

<p style = "font-size : 30px; color :#F2D742 ; font-family : 'Comic Sans MS'; text-align : center; background-color :#263049 ; border-radius: 5px 5px;"><strong> 4.1 Replace Missing Value of 'Rating' With average Rating in Each 'Install Category'</strong></p>


```python
# Step 1: Calculate the average rating for each 'Installs_category'
mean_ratings = df.groupby('Installs_category')['Rating'].mean()

# Step 2: Replace the missing values in the 'Rating' column
for index, row in df.iterrows():
    if pd.isnull(row['Rating']):
        installs_category = row['Installs_category']
        mean_rating = mean_ratings[installs_category]
        df.loc[index, 'Rating'] = mean_rating
```
